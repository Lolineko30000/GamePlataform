### Biblioteca y deseados

```python
def fechaAleatoria(inicio, final):
    if inicio>final:
        inicio, final = final, inicio
    # Calcular el rango de días entre la fecha de lanzamiento y la fecha actual
    diferencia_dias = final - inicio
    rango_dias = diferencia_dias.days

    # Generar un número aleatorio de días desde la fecha de lanzamiento hasta hoy
    dias_adquisicion = random.randint(0, rango_dias)

    # Sumar el número aleatorio de días a la fecha de lanzamiento para obtener la fecha de adquisición
    fecha_Aleatoria = inicio + timedelta(days=dias_adquisicion)
    return fecha_Aleatoria
```

```python
IdJuego1 = []
IdUsuario1 = []
ModoAdquisicion = []
FechaAdquisicion = []
Descuento = []
PrecioCompra = []

IdJuego2 = []
IdUsuario2 = []
FechaAgregado = []

n = Juego.shape[0]
# Obtener la fecha actual
actual = datetime.now()

for i in range(1000):
    comprados = []
    deseados = []
    for j in range(10):
        k = random.randint(0, n-1)
        [desc] = np.random.triangular(left=-.4, mode=.2, right=.8, size=1)
        if desc < 0: desc=0

        # Fecha de lanzamiento del juego en formato "AAAA-MM-DD" (como string) a datetime
        lanzamiento =  datetime.strptime(Juego.iloc[j]["FechaLanzamiento"], "%Y-%m-%d")
        
        if Usuario.iloc[i]["SaldoCartera"] >= Juego.iloc[k]["Precio"]*(1-desc):
            if k not in comprados:
                comprados.append(k)

                Usuario.at[i, "SaldoCartera"] = Usuario.at[i, "SaldoCartera"] - Juego.at[k, "Precio"] * (1 - desc)

                IdJuego1.append(k+1)
                IdUsuario1.append(i+1)
                ModoAdquisicion.append(0)
                FechaAdquisicion.append(fechaAleatoria(lanzamiento, actual).strftime("%Y-%m-%d"))
                Descuento.append(desc)
                PrecioCompra.append(Juego.iloc[k]["Precio"]*(1-desc))
        else:
            if k not in deseados:
                deseados.append(k)

                IdJuego2.append(k+1)
                IdUsuario2.append(i+1)
                FechaAgregado.append(lanzamiento)

datos_biblioteca = {
    "IdJuego": IdJuego1,
    "IdUsuario": IdUsuario1,
    "ModoAdquisicion": ModoAdquisicion,
    "FechaAdquisicion": FechaAdquisicion,
    "Descuento": Descuento,
    "PrecioCompra": PrecioCompra
}

datos_deseados = {
    "IdJuego": IdJuego2,
    "IdUsuario": IdUsuario2,
    "FechaAgregado": FechaAgregado,
}
                        
Biblioteca = pd.DataFrame(datos_biblioteca)
Deseado = pd.DataFrame(datos_deseados)
```

```python
IdJuego = []
IdUsuario = []
FechaInicio = []
FechaFin = []
TipoPartida = []


n = Avatar.shape[0]
actual = datetime.now()
for i in range(n):
    iJ = Biblioteca.iloc[i]["IdJuego"]
    iU = Biblioteca.iloc[i]["IdUsuario"]

    # Fecha de lanzamiento del juego en formato "AAAA-MM-DD" (como string) a datetime
    lanzamiento =  datetime.strptime(Juego.iloc[iJ-1]["FechaLanzamiento"], "%Y-%m-%d")

    m = random.randint(0, 60)
    for j in range(m):
        diaJuego = fechaAleatoria(lanzamiento, actual)
        hora = random.randint(0, 23)
        minuto = random.randint(0, 59)
        segundo = random.randint(0, 59)
        # Asignar la hora, minutos y segundos aleatorios a la fecha de adquisición
        fecha_juego = diaJuego.replace(hour=hora, minute=minuto, second=segundo)

        decisor = random.random()
        if decisor < 0.2:
            sJugados = random.randint(0, 1200)
        elif decisor < 0.99:
            [sJugados] = np.random.poisson(3000, 1)
        else:
            sJugados = pareto.rvs(.35, 4800)

        juego_final = fecha_juego + timedelta(seconds=int(sJugados))

        fecha_juego = fecha_juego.strftime("%Y-%m-%d %H:%M:%S")
        juego_final = juego_final.strftime("%Y-%m-%d %H:%M:%S")

        IdJuego.append(iJ)
        IdUsuario.append(iU)
        FechaInicio.append(fecha_juego)
        FechaFin.append(juego_final)
        TipoPartida.append(0)
    


datos_partidas = {
    "IdJuego": IdJuego,
    "IdUsuario": IdUsuario,
    "FechaInicio": FechaInicio,
    "FechaFin": FechaFin,
    "TipoPartida": TipoPartida
}

Partida = pd.DataFrame(datos_partidas)
Partida
```
